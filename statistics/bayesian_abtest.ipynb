```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

# Define the number of arms (webpage designs)
n_arms = 3

# Initialize observations and conversions for each arm with some small values to avoid division by zero
observations = [1] * n_arms  # Number of times each arm was shown
conversions = [1] * n_arms  # Number of conversions for each arm

def thompson_sampling(observations, conversions):
    samples = [beta(a=conversions[i]+1, b=observations[i]-conversions[i]+1).rvs() for i in range(n_arms)]
    return np.argmax(samples)

# Simulate showing webpages and observing conversions
np.random.seed(42)  # For reproducible results
for _ in range(1000):  # Assume 1000 visitors
    chosen_arm = thompson_sampling(observations, conversions)
    # Simulate conversion rate for simplicity; replace with real observation
    conversion_rate = [0.1, 0.15, 0.2]  # Hypothetical true conversion rates for each design
    conversion = np.random.rand() < conversion_rate[chosen_arm]
    observations[chosen_arm] += 1
    conversions[chosen_arm] += conversion

print("Final observed conversion rates:", [conversions[i] / observations[i] for i in range(n_arms)])
print("Number of times each arm was shown:", observations)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

def plot_beta_distribution(alpha, beta_param):
    # Generate a range of values between 0 and 1
    x = np.linspace(0, 1, 1000)
    
    # Calculate the PDF of the Beta distribution for each x value
    y = beta.pdf(x, alpha, beta_param)
    
    # Plot the PDF
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'Beta({alpha}, {beta_param})')
    
    # Labeling the plot
    plt.title('Beta Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    print( alpha/(alpha+beta_param))
    plt.show()

# Example usage: Plotting a Beta(2, 5) distribution
plot_beta_distribution(0.5, 0.8)

```
