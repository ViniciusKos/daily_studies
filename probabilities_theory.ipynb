```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
# Parameters
population_std = 30
sample_sizes = np.arange(10, 100001, 1000)  # Sample sizes from 10 to 10000
num_experiments = 100  # Number of experiments to average over

# Function to simulate experiments
def simulate_experiments(sample_sizes, population_mean, population_std, num_experiments):
    sample_means = []

    for size in sample_sizes:
        experiment_means = []
        for _ in range(num_experiments):
            sample = np.random.normal(population_mean, population_std, size)
            experiment_means.append(np.mean(sample))
        sample_means.append(np.mean(experiment_means))

    return sample_means

# Simulate experiments
sample_means = simulate_experiments(sample_sizes, population_mean, population_std, num_experiments)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, sample_means, label='Empirical Mean')
plt.axhline(y=population_mean, color='r', linestyle='-', label='True Population Mean')
plt.title('Law of Large Numbers Demonstration')
plt.xlabel('Sample Size')
plt.ylabel('Mean of Sample')
plt.legend()
plt.show()
```

```python
# Parameters for the new demonstration with an exponential distribution
population_mean = 50
lambda_param = 1 / population_mean  # Lambda parameter for the exponential distribution

# Function to generate samples from an exponential distribution and calculate means
def generate_exponential_sample_means(sample_size, num_samples, lambda_param):
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.exponential(1/lambda_param, sample_size)
        sample_means.append(np.mean(sample))

    return sample_means

# Generating sample means from the exponential distribution
exp_sample_means = generate_exponential_sample_means(30, 10000, lambda_param)

# Plotting the distribution of sample means from the exponential distribution
plt.figure(figsize=(10, 6))
plt.hist(exp_sample_means, bins=30, edgecolor='black', density=True)
plt.title('Distribution of Sample Means from Exponential Distribution (CLT Demonstration)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()

```
